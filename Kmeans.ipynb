```python
import numpy as np
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
cluster1 = np.random.normal(loc=0, scale=1, size=(100,2))
cluster2 = np.random.normal(loc=5, scale=1, size=(100,2))
```

```python
dataset = np.vstack([cluster1, cluster2])
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 42)
kmeans.fit(dataset)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dataset[:,0],dataset[:,1],c=labels, cmap = 'viridis', s=50, alpha=0.5)
plt.scatter(centroids[:,0],centroids[:,1],c='red', marker='X', s = 200, label = 'centroids')
plt.title('kmeans Clustering')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()
plt.show()
```
